Nạp các thư viện cần thiết.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Đọc bộ dữ liệu.

```python
url='https://raw.githubusercontent.com/vonguyenminhtu2504/Decision-Support-and-Business-Intelligence-Systems/main/all_data.csv'
sales_data = pd.read_csv(url)
sales_data.head()
```

Đếm số lượng các giá trị NULL (bị thiếu).

```python
null_values_count = sales_data.isnull().sum()
print(null_values_count)
```

Tính tỉ lệ phần trăm giá trị NULL so với toàn bộ dữ liệu.

```python
total_cells = np.product(sales_data.shape)
total_null_cells = null_values_count.sum()
null_percentage = (total_null_cells / total_cells) * 100
print("NULL percentage is: ", f"{null_percentage:.2f}%")
```

Vì tỉ lệ nhỏ nên có thể trực tiếp loại bỏ.

```python
sales_data = sales_data.dropna(how='all')
```

Kiểm tra lại số lượng các giá trị NULL.

```python
null_values_count = sales_data.isnull().sum()
print(null_values_count)
```

Loại bỏ lỗi.

```python
sales_data = sales_data[sales_data['Order Date'].str[0:2] != 'Or']
```

Đếm tổng số lượng đã bán ra của mỗi sản phẩm.

```python
sold_count = sales_data.groupby('Product')['Quantity Ordered'].count()
print(sold_count)
```

[link text](https:// [link text](https:// [link text](https://)))Chọn sản phẩm để xem xét. Nên chọn sản phẩm phổ biến để tăng độ chính xác.

```python
product_for_filtering = 'USB-C Charging Cable'
```

Tạo ma trận để tính tương quan giữa sản phẩm muốn xem xét và các sản phẩm còn lại.

```python
address_product_quantity = sales_data.pivot_table(index = 'Purchase Address', columns = 'Product', values = 'Quantity Ordered')
address_product_quantity.head()
```

Tìm ra khuôn mẫu của sản phẩm (cách mà mọi người mua nó).

```python
filtering_pattern = address_product_quantity[product_for_filtering]
filtering_pattern.head()
```

Tạo một ma trận mới và loại bỏ sản phẩm muốn xem xét ra khỏi ma trận này.

```python
address_product_quantity_for_filtering = address_product_quantity.drop(columns = product_for_filtering)
```

Tính hệ số tương quan giữa sản phẩm đang xem xét và các sản phẩm còn lại.

```python
product_like_pattern = address_product_quantity_for_filtering.corrwith(filtering_pattern)
corrwith_pattern = pd.DataFrame(product_like_pattern, columns = {'Correlation'})
corrwith_pattern.head()
```

Sắp xếp lại hệ số tương quan để tìm ra các sản phẩm có khuôn mẫu giống với sản phẩm đang xem xét nhất (là sản phẩm có hệ số tương quan lớn nhất, được sắp xếp để nằm trên cùng).

```python
corrwith_pattern = corrwith_pattern.sort_values('Correlation', ascending = False).head()
corrwith_pattern.head()
```

Chúng ta thấy có sự chênh lệch khá lớn giữa sản phẩm thứ nhất và sản phẩm thứ hai, nên ta sẽ bổ sung thêm số lượng đã bán ra của các sản phẩm này để phân tích thêm.

```python
corrwith_pattern_with_count = corrwith_pattern.join(sold_count)
print(corrwith_pattern_with_count)
```

Do chênh lệnh về số lượng bán ra nên dẫn đến sự chênh lệnh về hệ số tương quan. Để các đề xuất đảm bảo độ chính xác thì chúng ta sẽ quy định số lượng bán ra của sản phẩm đề xuất không được thấp hơn 1/4 số lượng bán ra của sản phẩm đang xem xét.

```python
final_dataframe = corrwith_pattern_with_count[corrwith_pattern_with_count['Quantity Ordered'] >= (sold_count[product_for_filtering] / 4)]
print(final_dataframe)
```

Vậy chúng ta có thể sử dụng những sản phẩm có hệ số tương quan và số lượng bán ra đủ cao trong danh sách trên để đề xuất cho những khách hàng đã từng mua sản phẩm chúng ta vừa xem xét.
